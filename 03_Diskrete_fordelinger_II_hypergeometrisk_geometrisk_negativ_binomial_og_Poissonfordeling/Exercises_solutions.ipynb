## Øvelse 1

In each of the following cases, determine which probability distribution to use and calculate the stated probability:

1. The worldwide average number of airplane crashes of commercial airlines is 3.5 crashes per month. Find the probability that there will be at least two crashes within the next month.
2. Suppose a batch of 100 items contains 6 that are defective. If you randomly select 10 items, what is the probability that you will select more than two defectives?
3. It is known that USB drives produced by a certain company will be defective with probability 0.01, independent of each other. The company sells the USBs in packages of size 10. If someone buys 3 packages, what is the probability that one or more of the packages contains at least one defective USB? (Two-step problem).

````python
# 1) Poisson: X ~ Poisson(lambda=3.5), find P(X>=2)
lam = 3.5
p1 = 1 - poisson.cdf(1, lam)

# 2) Hypergeometrisk: M=100 total, n=6 defective, N=10 draws, find P(X>2)
M, n, N = 100, 6, 10
p2 = 1 - hypergeom.cdf(2, M, n, N)

# 3) To-trins binomial
# Trin 1: sandsynlighed for mindst 1 defekt i en pakke med 10 USB
p_defect_usb = 0.01
p_pkg_bad = 1 - binom.pmf(0, 10, p_defect_usb)

# Trin 2: 3 pakker, sandsynlighed for at mindst én pakke er "bad"
p3 = 1 - binom.pmf(0, 3, p_pkg_bad)

print(f"1) Poisson, P(X>=2): {p1:.6f}")
print(f"2) Hypergeometric, P(X>2): {p2:.6f}")
print(f"3) Two-step binomial, P(at least one bad package): {p3:.6f}")
````

### WolframAlpha

**1) Poisson**

```

**2) Hypergeometrisk**

```

```

**3) To-trins model**

Trin 1:

```

Trin 2 (brug resultatet fra trin 1 som sandsynlighed):

## Øvelse 2

A player of a video game is confronted with a series of opponents and has an 80% probability of defeating each one. Success with any opponent is independent of previous encounters. Until defeated, the player continues to contest opponents.

1. What is the probability mass function of the number of opponents contested in a game?
2. What is the probability that a player defeats at least two opponents in a game?
3. What is the expected number of opponents contested in a game?
4. What is the probability that a player contests four or more opponents in a game?
5. What is the expected number of game plays until a player contests four or more opponents?

````python
# X = antal opponents contested indtil første nederlag
# P(vinde) = 0.8, P(nederlag) = q = 0.2
p_win = 0.8
q_lose = 0.2

# 2) Defeats at least two opponents: vinder de første to
p2 = p_win**2

# 3) E[X] for geom med "succes" = nederlag (q_lose)
EX = 1 / q_lose

# 4) Contest four or more opponents: overlever første 3
p4 = p_win**3

# 5) Antal game plays indtil første gang eventen (X>=4) sker
# Y ~ Geometric(p4), E[Y] = 1/p4
EY = 1 / p4

print("1) PMF: P(X=k) = 0.8^(k-1) * 0.2,  k=1,2,3,...")
print(f"2) P(defeats at least two) = {p2:.6f}")
print(f"3) E[X] = {EX:.6f}")
print(f"4) P(X>=4) = {p4:.6f}")
print(f"5) E[Y] = {EY:.6f}")
````

### WolframAlpha

Hvis vi sætter $X$ som antal opponents contested indtil første nederlag:

- **PMF (fx for $k=4$)**

```
P(X=4) GeometricDistribution(0.2)
```

(Bemærk: WolframAlpha `GeometricDistribution(p)` tæller failures før første succes. Her bruges primært Python-resultaterne ovenfor til fortolkning.)

- **P(defeats at least two opponents)**

```
0.8^2
```

- **P(X>=4)**

```
0.8^3
```

- **E[Y] hvor Y er antal game plays indtil X>=4**

```
1/(0.8^3)
```

## Øvelse 3

High flows result in the closure of a causeway. From past records, the road is closed for this reason on 10 days during a 20-year period. At an adjoining village, there is concern about the closure of the causeway because it provides the only access. The villagers assume that the probability of a closure of the road for more than one day during a year is less than 0.10. Is this correct? Show using the Poisson distribution.

````python
# 10 closure-days over 20 years -> lambda per year = 0.5
lam = 10 / 20

# P(X > 1) = 1 - P(X <= 1)
p = 1 - poisson.cdf(1, lam)

print(f"lambda per year = {lam:.3f}")
print(f"P(X>1) = {p:.6f}")
print(f"Is P(X>1) < 0.10? {'Yes' if p < 0.10 else 'No'}")
````

### WolframAlpha

```
P(X>1) PoissonDistribution(0.5)
```

## Øvelse 5

A startup's sales team cold-calls potential customers. Each call results in a sale with probability $p = 0.30$, independently of other calls. Let $X$ be the number of calls needed until the team has made $r=4$ sales.

1. What is the probability that the team needs at most 12 calls to make 4 sales?
2. What is the probability that the 4th sale occurs exactly on call 12?
3. Find $E[X]$ and $\mathrm{Var}(X)$.
4. What is the probability that the team needs more than 20 calls to make 4 sales?

````python
# Negativ binomial i SciPy: nbinom(r, p) returnerer antal failures før r-te succes
r, p = 4, 0.30

# X = antal calls indtil r succeser
# failures F = X - r

# 1) P(X<=12) -> P(F<=8)
p1 = nbinom.cdf(12 - r, r, p)

# 2) P(X=12) -> P(F=8)
p2 = nbinom.pmf(12 - r, r, p)

# 3) E[X], Var(X)
EX = r / p
VarX = r * (1 - p) / (p**2)

# 4) P(X>20) -> P(F>16)
p4 = nbinom.sf(20 - r, r, p)

print(f"1) P(X<=12) = {p1:.6f}")
print(f"2) P(X=12)  = {p2:.6f}")
print(f"3) E[X] = {EX:.6f}, Var(X) = {VarX:.6f}")
print(f"4) P(X>20)  = {p4:.6f}")
````

### WolframAlpha

`NegativeBinomialDistribution(4, 0.30)` bruger variablen som antal **failures** før 4. succes.
Sæt derfor $F = X - 4$.

- **$P(X\leq 12)=P(F\leq 8)$**

```
P(F<=8) NegativeBinomialDistribution(4, 0.30)
```

- **$P(X=12)=P(F=8)$**

```
P(F=8) NegativeBinomialDistribution(4, 0.30)
```

- **Middelværdi og varians for failures $F$**

```
mean NegativeBinomialDistribution(4, 0.30)
```

```

```

For calls gælder: $X=F+4$, så $E[X]=E[F]+4$ og $\mathrm{Var}(X)=\mathrm{Var}(F)$.

- **$P(X>20)=P(F>16)$**

```
P(F>16) NegativeBinomialDistribution(4, 0.30)
```

## Øvelse 6

Astronomers treat the number of stars in a given volume of space as a Poisson random variable. The density in the Milky Way Galaxy in the vicinity of our solar system is one star per 16 cubic light-years.

1. What is the probability of no stars in 16 cubic light-years?
2. What is the probability of two or more stars in 16 cubic light-years?
3. How many cubic light-years of space must be studied so that the probability of one or more stars exceeds 0.95?

````python
# Densitet: 1 stjerne pr. 16 kubik-lysår
# For volumen V er lambda = V/16

lam_16 = 16 / 16  # =1

# 1) P(X=0) for V=16
p1 = poisson.pmf(0, lam_16)

# 2) P(X>=2) for V=16
p2 = 1 - poisson.cdf(1, lam_16)

# 3) Find V så P(X>=1)>0.95
# 1 - exp(-V/16) > 0.95  =>  exp(-V/16) < 0.05  =>  V > -16 ln(0.05)
V_min = -16 * np.log(0.05)

print(f"1) P(X=0 | V=16)  = {p1:.6f}")
print(f"2) P(X>=2 | V=16) = {p2:.6f}")
print(f"3) Required volume V > {V_min:.6f} cubic light-years")
````

- **$P(X\geq 2)$ for 16 cubic light-years**
