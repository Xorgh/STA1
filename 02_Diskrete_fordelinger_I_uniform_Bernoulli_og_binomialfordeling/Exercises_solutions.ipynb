## Øvelse 1

A computer system uses passwords that are exactly six characters and each character is one of the 26 letters (a–z) or 10 integers (0–9). Suppose that 10,000 users of the system have unique passwords. A hacker randomly selects (with replacement) 100,000 passwords from the potential set, and a match to a user's password is called a hit.

1. What is the distribution of the number of hits?
2. What is the probability of no hits?
3. What are the mean and variance of the number of hits?

````python
n = 10**5  # antal forsøg
p = 10**4 / (36**6)  # sandsynlighed for hit (10000 brugere / 36^6 mulige passwords)

# 1. Fordeling: X ~ Binomial(n, p)
print(f"1. Fordeling: X ~ Binomial(n={n}, p={p:.6e})")

# 2. P(X=0)
p_no_hits = binom.pmf(0, n, p)
print(f"2. P(X=0) = {p_no_hits:.4f}")

# 3. Middelværdi (Expected Value) og varians
mean = n * p
var = n * p * (1 - p)
print(f"3. E[X] = {mean:.4f}, Var(X) = {var:.4f}")
````

### WolframAlpha

- **PMF**:

```
P(X=0) BinomialDistribution(100000, 10000/36^6)
```

- **Middelværdi (Expected Value)**:

```
mean BinomialDistribution(100000, 10000/36^6)
```

- **Varians**:

```
variance BinomialDistribution(100000, 10000/36^6)
```

## Øvelse 2

Because all airline passengers do not show up for their reserved seat, an airline sells 125 tickets for a flight that holds only 120 passengers. The probability that a passenger does not show up is 0.10, and the passengers behave independently.

1. What is the probability that every passenger who shows up can take the flight?
2. What is the probability that the flight departs with empty seats?

````python
n = 125  # antal solgte billetter
p = 0.10  # sandsynlighed for at møde op
# X = antal der IKKE møder op ~ Binomial(125, 0.10)

# 1. Alle der møder op kan tage flyet: mindst 5 møder ikke op
#    X >= 5
p_all_board = 1 - binom.cdf(4, n, 0.10)
print(f"1. P(alle kan tage flyet) = P(X ≥ 5) = {p_all_board:.4f}")

# 2. Flyet letter med tomme pladser: mere end 5 møder ikke op
#    X > 5  <=>  X >= 6
p_empty_seats = 1 - binom.cdf(5, n, 0.10)
print(f"2. P(tomme pladser) = P(X ≥ 6) = {p_empty_seats:.4f}")
````

### WolframAlpha

- **Sandsynlighed for at alle kan komme med**:

```
P(X>=5) BinomialDistribution(125, 0.10)
```

- **Sandsynlighed for tomme sæder**:

```
P(X>=6) BinomialDistribution(125, 0.10)
```

## Øvelse 3

A congested computer network has a 1% chance of losing a data packet that must be resent, and packet losses are independent events. An e-mail message requires 100 packets.

1. What is the probability that at least one packet is resent?
2. What is the probability that two or more packets are resent?
3. What are the mean and standard deviation of the number of packets that are resent?
4. If there are 10 messages and each contains 100 packets, what is the probability that at least one message requires that two or more packets be resent?

````python
n = 100  # antal pakker per besked
p = 0.01  # sandsynlighed for at miste en pakke
# X = antal pakker der sendes igen ~ Binomial(100, 0.01)

# 1. P(X >= 1)
p_at_least_one = 1 - binom.pmf(0, n, p)
print(f"1. P(X ≥ 1) = {p_at_least_one:.4f}")

# 2. P(X >= 2)
p_two_or_more = 1 - binom.cdf(1, n, p)
print(f"2. P(X ≥ 2) = {p_two_or_more:.4f}")

# 3. Middelværdi (Expected Value) og standardafvigelse
mean = n * p
std = np.sqrt(n * p * (1 - p))
print(f"3. E[X] = {mean}, σ = {std:.3f}")

# 4. Y = antal besked'er (ud af 10) der kræver 2+ genafsending
#    Y ~ Binomial(10, p_two_or_more)
#    P(Y >= 1) = 1 - P(Y=0)
p_at_least_one_msg = 1 - binom.pmf(0, 10, p_two_or_more)
print(f"4. P(mindst én besked kræver 2+ genafsending) = {p_at_least_one_msg:.4f}")
````

### WolframAlpha

- **$P(X \geq 1)$**:

```
P(X>=1) BinomialDistribution(100, 0.01)
```

- **$P(X \geq 2)$**:

```
P(X>=2) BinomialDistribution(100, 0.01)
```

- **Middelværdi (Expected Value)**:

```
mean BinomialDistribution(100, 0.01)
```

- **Standardafvigelse**:

```
standard deviation BinomialDistribution(100, 0.01)
```

For spørgsmål 4: Først beregn $P(X \geq 2)$ ovenfor ($\approx 0.2642$). Derefter:

```
P(Y>=1) BinomialDistribution(10, 0.2642)
```

## Øvelse 4

The probability that a patient recovers from a rare blood disease is 0.4. If 15 people are known to have contracted this disease, what is the probability that:

1. At least 10 survive
2. From 3 to 8 survive
3. Exactly 5 survive
4. Find the mean and variance.

````python
n = 15
p = 0.4
# X = antal der overlever ~ Binomial(15, 0.4)

# 1. P(X >= 10) = 1 - P(X <= 9)
p_at_least_10 = 1 - binom.cdf(9, n, p)
print(f"1. P(X ≥ 10) = {p_at_least_10:.4f}")

# 2. P(3 <= X <= 8) = P(X <= 8) - P(X <= 2)
p_3_to_8 = binom.cdf(8, n, p) - binom.cdf(2, n, p)
print(f"2. P(3 ≤ X ≤ 8) = {p_3_to_8:.4f}")

# 3. P(X = 5)
p_exactly_5 = binom.pmf(5, n, p)
print(f"3. P(X = 5) = {p_exactly_5:.4f}")

# 4. Middelværdi (Expected Value) og varians
mean = n * p
var = n * p * (1 - p)
print(f"4. E[X] = {mean}, Var(X) = {var}")
````

### WolframAlpha

- **$P(X \geq 10)$**:

```
P(X>=10) BinomialDistribution(15, 0.4)
```

- **$P(3 \leq X \leq 8)$**:

```
P(3<=X<=8) BinomialDistribution(15, 0.4)
```

- **$P(X = 5)$**:

```
P(X=5) BinomialDistribution(15, 0.4)
```

- **Middelværdi (Expected Value)**:

```
mean BinomialDistribution(15, 0.4)
```

- **Varians**:

```
variance BinomialDistribution(15, 0.4)
```

## Øvelse 5

Let $X$ and $Y$ be two independent discrete random variables with custom PMFs. Find:

1. $P(X<2 \text{ and } Y<2)$
2. $P(X \ge 2 \text{ or } Y \ge 2)$
3. $P(X \ge 2 \mid Y \ge 2)$
4. $P(X < Y)$

Bemærk: Det gir ikke så meget mening at lave denne opgave i Python. Det er altså nemmere med pen og papir.

````python
# PMF for X: P(X=0)=1/10, P(X=1)=2/10, P(X=2)=3/10, P(X=3)=4/10
px = {0: 1/10, 1: 2/10, 2: 3/10, 3: 4/10}
# PMF for Y: P(Y=0)=1/5, P(Y=1)=1/5, P(Y=2)=2/5, P(Y=3)=1/5
py = {0: 1/5, 1: 1/5, 2: 2/5, 3: 1/5}

# 1. P(X<2 and Y<2) = P(X=0 or X=1) * P(Y=0 or Y=1)  (uafhængige)
p_x_lt_2 = px[0] + px[1]
p_y_lt_2 = py[0] + py[1]
p1 = p_x_lt_2 * p_y_lt_2
print(f"1. P(X<2 and Y<2) = {p1:.2f}")

# 2. P(X>=2 or Y>=2) = 1 - P(X<2 and Y<2)
p2 = 1 - p1
print(f"2. P(X≥2 or Y≥2) = {p2:.2f}")

# 3. P(X>=2 | Y>=2) = P(X>=2 and Y>=2) / P(Y>=2)
p_x_ge_2 = px[2] + px[3]
p_y_ge_2 = py[2] + py[3]
p_x_ge_2_and_y_ge_2 = p_x_ge_2 * p_y_ge_2
p3 = p_x_ge_2_and_y_ge_2 / p_y_ge_2
print(f"3. P(X≥2 | Y≥2) = {p3:.2f}")

# 4. P(X < Y) = sum over (x,y) where x < y: P(X=x)*P(Y=y)
p4 = 0
for x in px:
    for y in py:
        if x < y:
            p4 += px[x] * py[y]
print(f"4. P(X < Y) = {p4:.2f}")
````

### Kombination: Pen og papir + WolframAlpha

Øvelse 5 bruger brugerdefinerede PMF'er, så man kan ikke bruge `X ~ Binomial(...)` direkte. En god måde er: **papir til strukturen**, **WolframAlpha til talregning**.

**1. $P(X<2 \text{ and } Y<2)$**  
Uafhængighed giver: $P(X<2) \cdot P(Y<2) = (\frac{1}{10} + \frac{2}{10}) \cdot (\frac{1}{5} + \frac{1}{5}) = \frac{3}{10} \cdot \frac{2}{5}$. Beregn i WolframAlpha:

```
(1/10 + 2/10) * (1/5 + 1/5)
```

**2. $P(X \ge 2 \text{ or } Y \ge 2)$**  
Komplementregel: $1 - P(X<2 \text{ and } Y<2)$. Brug resultatet fra 1:

```
1 - 3/10 * 2/5
```

**3. $P(X \ge 2 \mid Y \ge 2)$**  
Ved uafhængighed: $P(X \ge 2 \mid Y \ge 2) = P(X \ge 2) = P(X=2) + P(X=3) = \frac{3}{10} + \frac{4}{10}$:

```
(3/10 + 4/10)
```

**4. $P(X < Y)$**  
Sum over par $(x,y)$ hvor $x < y$: $P(X=x) \cdot P(Y=y)$. På papir finder du de 6 led $(0,1)$, $(0,2)$, $(0,3)$, $(1,2)$, $(1,3)$, $(2,3)$. I WolframAlpha:

```
1/10*1/5 + 1/10*2/5 + 1/10*1/5 + 2/10*2/5 + 2/10*1/5 + 3/10*1/5
```

eller kortere:

```
(1*1 + 1*2 + 1*1 + 2*2 + 2*1 + 3*1) / (10*5)
```
